# Chap 1

## iris

- 返ってくるのは dict like な Bunch Object

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
iris_dataset.keys()
```

### DESCRには説明書きが入ってる

```python
print(iris_dataset['DESCR'][:200])
```

### train_test_splitでTestとTrainにデータを分割する

- split data set into 2 parts
 - training data for building models
 - test data for testing how model is acurate
 
- X is data, y is label

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print(X_train.shape) #75%のデータ
print(X_test.shape) # 25%のデータ
```

### データを見る

#### DFにする

```python
import pandas as pd
iris_df = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
iris_df
```

#### Scatter Matrix

```python
%matplotlib inline
from pandas.plotting import scatter_matrix
scatter_matrix(iris_df, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=8)
```

### k-nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

#### making predictions

検証したいデータを作る

```python
import numpy as np
X_new = np.array([[5, 2.9, 1, 0.2]])
print(X_new.shape)
```

class 0, setosa と言う種類だとわかる

```python
prediction = knn.predict(X_new)
print(prediction)
print(iris_dataset['target_names'][prediction])
```
