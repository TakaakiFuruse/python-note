### train_test_splitでTestとTrainにデータを分割する

- split data set into 2 parts
 - training data for building models
 - test data for testing how model is acurate
 
- X is data, y is label

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

## generalization, overfitting, underfitting

- generalization
  - model can make accurate prediction 
    - the model can generalize from traning set to test set.
- overfitting
  - fit to trainnig data 
  - but does not works on testing data
- underfitting
 - over generalized the matter
 
- Complexed model => Overfit model to traning data, not works on new data

### 使ってみる

データを投入して、分割

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

関数に入れる。隣は3つ

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
clf
```

calculate Nearest Neighbor

```python
clf.fit(X_train, y_train)
```

make a prediction from test data

```python
clf.predict(X_test)
```

check score 

```python
clf.score(X_test, y_test)
```

### Decision boundry

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))
for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
    axes[0].legend(loc=3)
```
